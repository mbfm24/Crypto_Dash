```python
# Import the required libraries and dependencies
import numpy as np
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Import the Bitcoin data by reading in the CSV file and setting the DatetimeIndex 
bitcoin_csv = pd.read_csv(
    Path('clean_data/Bitcoin_clean.csv'),
    index_col="date",
    parse_dates=True,
    infer_datetime_format=True
)

bitcoin_csv = bitcoin_csv.rename(columns={'daily_return': 'bitcoin_daily_return'})
# View the first 5 rows of the DataFrame
bitcoin_csv.head().dropna()
```

```python
# Import the Chainlink data by reading in the CSV file and setting the DatetimeIndex 
chainlink_csv = pd.read_csv(
    Path('clean_data/Chainlink_clean.csv'),
    index_col="date",
    parse_dates=True,
    infer_datetime_format=True
)
chainlink_csv = chainlink_csv.rename(columns={'daily_return': 'chainlink_daily_return'})
# View the first 5 rows of the DataFrame
chainlink_csv.head().dropna()
```

```python
# Import the Doge data by reading in the CSV file and setting the DatetimeIndex 
doge_csv = pd.read_csv(
    Path('clean_data/Doge_clean.csv'),
    index_col="date",
    parse_dates=True,
    infer_datetime_format=True
)
doge_csv = doge_csv.rename(columns={'daily_return': 'doge_daily_return'})
# View the first 5 rows of the DataFrame
doge_csv.head().dropna()
```

```python
# Import the Ethereum data by reading in the CSV file and setting the DatetimeIndex 
ethereum_csv = pd.read_csv(
    Path('clean_data/Ethereum_clean.csv'),
    index_col="date",
    parse_dates=True,
    infer_datetime_format=True
)
ethereum_csv = ethereum_csv.rename(columns={'Unnamed: 5': 'ethereum_daily_return'})
# View the first 5 rows of the DataFrame
ethereum_csv.head().dropna()
```

```python
# Import the Litecoin data by reading in the CSV file and setting the DatetimeIndex 
litecoin_csv = pd.read_csv(
    Path('clean_data/Litecoin_clean.csv'),
    index_col="date",
    parse_dates=True,
    infer_datetime_format=True
)
litecoin_csv = litecoin_csv.rename(columns={'daily_return': 'litecoin_daily_return'})
# View the first 5 rows of the DataFrame
litecoin_csv.head().dropna()
```

```python
# Import the ZCash data by reading in the CSV file and setting the DatetimeIndex 
zcash_csv = pd.read_csv(
    Path('clean_data/Zcash_clean.csv'),
    index_col="date",
    parse_dates=True,
    infer_datetime_format=True
)
zcash_csv = zcash_csv.rename(columns={'daily_return': 'zcash_daily_return'})
# View the first 5 rows of the DataFrame
zcash_csv.head().dropna()
```

```python
# Concatenate cryto dataframes
crypto_coins_df = pd.concat([bitcoin_csv, chainlink_csv, doge_csv, litecoin_csv, zcash_csv], axis=1).dropna()

crypto_coins_df.head()
```

```python
# Create daily returns dataframe
daily_returns_df = crypto_coins_df[['bitcoin_daily_return', 'chainlink_daily_return', 'doge_daily_return', 'litecoin_daily_return', 'zcash_daily_return']].copy()

daily_returns_df.head()
```

```python
# Plot the daily return data 
daily_returns_df.plot(figsize = (14,7), title = "Daily Returns - Cryto Coins ")
```

```python
# Calculate the cumulative returns
cumulative_returns = (1 + daily_returns_df).cumprod()

cumulative_returns.head()
```

```python
# Visualize the cumulative returns using the Pandas plot function
cumulative_returns.plot(figsize = (14, 7), title = "Cumulative Returns - Crypto Coins ")
```

```python
# Use the daily return data to create box plots to visualize the volatility
daily_returns_df.plot.box(figsize = (14,7), title = "Box Plot of Daily Returns - Crypto Coins")
```

```python
# Calculate and sort the standard deviation for Crypto Coins
standard_deviation = daily_returns_df.std()

standard_deviation_sorted = standard_deviation.sort_values()

standard_deviation_sorted
```

```python
# Calculate and sort the annualized standard deviation (365 trading days) of the Crypto Coins
annualized_standard_deviation = standard_deviation * np.sqrt(365)

annualized_standard_deviation_sorted = annualized_standard_deviation.sort_values()

annualized_standard_deviation_sorted
```

```python
# The daily returns DataFrame and a 20-day rolling window, 
# plot the rolling standard deviation of the Crypto Coins
ax = daily_returns_df.plot(figsize=(14,7), title="Crypto Coins - 20-Day Rolling Average")

daily_returns_df.rolling(window=20).std().plot(ax=ax)

ax.legend(["Daily Returns", "20-Day Rolling Average"])
```

```python
# Calculate the annual average return data for the for the Crypto Coins (365 trading days)
annual_average_returns = daily_returns_df.mean() * 365

annual_average_returns_sorted = annual_average_returns.sort_values()

annual_average_returns_sorted
```

```python
# Calculate the annualized Sharpe Ratios for each of the Crypto Coins
sharpe_ratios = annual_average_returns / annualized_standard_deviation

sharpe_ratios.sort_values()
```

```python
# Plot the Sharpe ratios as a bar chart
sharpe_ratios.plot.bar(figsize=(14,7), title="Crypto Coins - Sharpe Ratios")
```
